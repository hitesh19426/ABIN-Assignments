```python
import csv
import random
import statistics
```

```python
file1 = open("SRP026387_scaledcounts.csv", "r")
reader1 = csv.reader(file1)
data = []
for row in reader1:
    data.append(row)
# print(data)

file1.close()
```

```python
allmeans = []
pivalues = []
for i in range(1, len(data)):
    rowi = list(map(int, data[i][1:]))

    for j in range(0, 20):
        random.shuffle(rowi)
        # print(rowi, end='\n')

        normal = rowi[0:6]
        tumer = rowi[6:]

        mean1 = statistics.mean(normal)
        mean2 = statistics.mean(tumer)
        sigma1 = statistics.median(normal)
        sigma2 = statistics.median(tumer)

        pivalue = (abs(mean1 - mean2)+1)/(sigma1+sigma2+1)
        absmean = abs(mean1 - mean2)
        # print(absmean)

        allmeans.append(absmean)
        pivalues.append(pivalue)

# print(allmeans)
print("pivalues = \n")
print(pivalues)
```
